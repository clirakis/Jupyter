```python
# includes
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.constants as const
```

```python
def Blackbody(F, T):
    """@brief - return the blackbody Energy as a function of frequency and temperature
    @param F - Frequency in Hz
    @param T - Temperature in Kelvin

    Boltzmann constant is Joules/Kelvin
    c is meters/sec
    h is Joule/Hertz
    @return power in Watts/m3
    Units don't come out right unless F^5, replicates curve. 
    """
    part1 = 8*const.pi*const.h*np.power(F,5.0)/np.power(const.c,3.0)
    part2 = 1.0/(np.exp(const.h*F/(const.k*T))-1)
    #print("part1 ", part1, " part 2 ", part2)
    rv = part1 * part2
    return rv
```

```python
lam = np.arange(100.0,2000.0,50.0)
lam = lam * 1.0e-9
f = const.c/lam
T = 5000.0     # star
E = Blackbody(f,T)
plt.plot(lam*1.0e9, E)
#print(E)
```
