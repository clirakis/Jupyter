```python
def GForce(mass, distance):
    """
    Newton's equation
    @param mass - mass in kg
    @param distance - distance from primary mass (M_earth) in meters
    """
    M_earth = 5.9722e+24 # kg 
    R_earth = 6378100.0  # meters
    r = distance + R_earth
    Force = konst.G * M_earth * mass/np.power(r, 2.0)
    return Force
```

```python
# Plot some object and the force on it as a funciton of distance
m = 1 # kg at earth surface. 
r = np.arange(1,2000e3, 10) # max out at LEO
F = GForce(m, r)
#print("G: ", konst.G)
acceleration = konst.G * 5.9722e+24/np.power(6378100,2.0)
print("g: ", acceleration)
print("test: ", GForce(1,0))
plt.plot(r,F)
plt.grid()
```
