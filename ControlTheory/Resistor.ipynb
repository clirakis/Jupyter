```python
def Theta(t, t0):
    index = np.where(t>t0)
    rv = np.zeros(len(t))
    rv[index] = 1
    return rv
```

```python
dt = 0.1  # time step in seconds
R  = 1000 # Ohms
t0 = 0.2  # where the theta function turns on 
# E = IR
I0 = 1.0e-3 # maximum Amps
t  =  np.arange(0.0, 0.5, dt)
I = I0 * Theta(t, 0.1)
E = I * R
plt.plot(t,E)
```
