```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.constants as const
```

```python
def BasicRC(R,C, t):
    """
    @param resistance in ohms
    @param C capicance in farads
    @param t
    @return expected voltage
    """
    tau = 1.0/(R*C)
    return np.exp(-t * tau), tau
```

```python
t = np.arange(0,100,0.1)
C = 400e-6 # farads
R = 10e3   # ohms
y, tau = BasicRC(R,C,t)
plt.plot(t,y)
plt.grid()
plt.xlabel('time')
plt.ylabel('volts')
print('Tau:', 1.0/tau)
```
