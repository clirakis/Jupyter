```python
Path     = "CTCN01085"
Year     = "2024"
Mon      = "Mar"
Prefix   = "CTCN01085_DL"
Suffix   = ".csv"
StartDay = 1
EndDay   = 32

# read the first day and create the base dataframe. 
Day    = "{:02d}".format(StartDay)
# Fully form filename
Filename = Path + '/' + Year + '/' + Mon + '/' + Prefix + Year + '_' + Mon + '_' + Day + Suffix
df = pd.read_csv(Filename)
# Now loop over the remaining days in the directory. 
for nday in range(StartDay+1,EndDay):
    Day    = "{:02d}".format(nday)
    Filename = Path + '/' + Year + '/' + Mon + '/' + Prefix + Year + '_' + Mon + '_' + Day + Suffix
    print('Processing: ',Filename)
    df2 = pd.read_csv(Filename)
    # Append this to the existing dataframe.
    df = pd.concat([df,df2])
    
#df.head()   # uncomment if you want to see what is in the head.
# convert the time column to datetime
df['dd:mm:yyyy hh:mm:ss'] = pd.to_datetime(df['dd:mm:yyyy hh:mm:ss'],dayfirst=True)
```

```python
# Plot the primary and Secondary loop temperatures
df.plot(x='dd:mm:yyyy hh:mm:ss',y=['T1','T5'])
plt.grid()
plt.title('Primary loop supply and return temperatures in C.')
```

```python
# Plot the  Secondary loop temperatures
df.plot(x='dd:mm:yyyy hh:mm:ss',y=['T2','T4'])
plt.grid()
plt.title('Secondary loop supply and return temperatures in C.')
```

```python
# Ambient conditions
df.plot(x='dd:mm:yyyy hh:mm:ss',y=['RH','DW'])
plt.grid()
plt.title('Ambient conditions')
```

```python
# Plot the Secondary loop parameters
df.plot(x='dd:mm:yyyy hh:mm:ss',y=['Sec Flow','Duty'])
plt.grid()
plt.title('Seconary loop parameters')
```

```python
# Plot the Secondary loop parameters
df.plot(x='dd:mm:yyyy hh:mm:ss',y= 'Duty')
plt.grid()
plt.title('Seconary loop Duty')
```
