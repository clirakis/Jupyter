SWR is defined as $VSWR = {1+|S_{11}|\over{1-|S_{11}|}}$ <br> 
For me the first thing to think about is the maximum power transfered only occurs when the source and sink impedence are the same. <br> 
<img src="Pic/Thev.png" /> <br>
Looking at the two circuits above a current source on the left and a voltage source on the right. In all cases 
there are source impedences associated with the source. On the left and right R1 is the source impedence. Thevin's Therom states that a current source with an internal impedance R1 is equal to a current source of the same impedance R1. Therefore left and right are equivalent circuits. Lets look at R2 now. R2 can be thought of as the load. We can measure the voltage across R2 as a direct indicator of the power transfered. <br> 
L.H.S. : $V_{R_{2}} = IR_{2}$ <br> 
R.H.S. : $V_{R_{2}} = IR_{2}$  but $I = {V\over{R_{1} + R_{2}}}$ therefore $V_{R_{2}} = {VR_{2}\over{R_{1} + R_{2}}}$<br>
Lets look at the voltage across the terminals of $R_{2}$ for a fixed R1. We will nomalize the plot to the source 
voltage. 

```python
R1 = 100. # ohms
R2 = np.arange(1,200,1)
VSource = 1.0
VR2 = VSource/(1.0+R2/R1)
plt.plot(R2, VR2)
plt.grid()
```

This just demonstrates we have a voltage divider. What about the power? $P=IE=I^2R = {V^2\over{R}}$

```python
P=np.power(VR2,2.0)/R2
plt.plot(R2,P)
plt.grid()
```
