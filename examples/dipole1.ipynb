# Dipole experiments

```python
import numpy as np
import matplotlib.pyplot as plt
import math
#
# https://matplotlib.org/stable/gallery/pie_and_polar_charts/polar_legend.html#sphx-glr-gallery-pie-and-polar-charts-polar-legend-py
# This is just the X-Y response. 
```

```python
# , arrange is (start,stop,step,dtype)
x = np.arange(-1.0, 1.0, 0.01)  # A single cycle
theta = 2.0 * np.pi * x
#
# Matplotlib examples: https://pythonguides.com/matplotlib-plot-numpy-array/
# In this example we will make a figure with 2 panes.
# some examples can be found here: 
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
#
# Dipole response
r = 1.0 - pow(np.cos(theta),2.0)

# Single plot
fig = plt.figure()
ax = fig.add_subplot(projection="polar", facecolor="lightgoldenrodyellow")

ax.plot(theta, r, color="tab:orange", lw=3, label="dipole")
ax.tick_params(grid_color="palegoldenrod")

plt.show()
```
